```python
import numpy as np

# Generate a time-domain signal
time = np.linspace(0, 1, 1000)
signal = np.sin(2 * np.pi * 10 * time) + 0.5 * np.sin(2 * np.pi * 25 * time)

# Compute the DFT using numpy.fft.fft()
frequency = np.fft.fftfreq(len(signal), 1 / len(time))
spectrum = np.abs(np.fft.fft(signal))

# Print the frequency and spectrum
for freq, amp in zip(frequency, spectrum):
    print(f"Frequency: {freq:.2f} Hz, Amplitude: {amp:.2f}")
```
