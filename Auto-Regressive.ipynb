```python
import numpy as np
import matplotlib.pyplot as plt

# Generate an auto-regressive process
order = 10
time = np.arange(0, 1000)
noise = np.random.normal(0, 1, len(time))
signal = np.zeros_like(time, dtype=np.float32)
for i in range(order, len(time)):
    signal[i] = 0.8 * signal[i-1] + 0.2 * signal[i-2] + noise[i]

# Plot the auto-regressive process
plt.plot(time, signal)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Auto-Regressive Process')
plt.show()
```
