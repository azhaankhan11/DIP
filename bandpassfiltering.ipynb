```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a noisy signal
t = np.linspace(0, 1, 1000)
signal = np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * 10 * t) + np.random.normal(0, 0.5, len(t))

# Compute the FFT
fft = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(signal), 1 / len(t))

# Define the frequency range for bandpass filtering
lower_freq = 4
upper_freq = 6

# Apply bandpass filtering
fft[(frequencies < lower_freq) | (frequencies > upper_freq)] = 0

# Compute the inverse FFT
filtered_signal = np.fft.ifft(fft)

# Plot the original and filtered signals
plt.plot(t, signal, label='Original Signal')
plt.plot(t, filtered_signal, label='Filtered Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Bandpass Filtering')
plt.legend()
plt.show()
```
