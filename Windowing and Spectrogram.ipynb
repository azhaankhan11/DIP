```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a time-domain signal
time = np.linspace(0, 1, 1000)
signal = np.sin(2 * np.pi * 10 * time) + 0.5 * np.sin(2 * np.pi * 25 * time)

# Compute the spectrogram using windowing
window_size = 128
overlap = 0.5
freq, time, spectrogram = plt.specgram(signal, NFFT=window_size, Fs=1000, noverlap=int(window_size * overlap))

# Plot the spectrogram
plt.xlabel('Time')
plt.ylabel('Frequency (Hz)')
plt.title('Spectrogram')
plt.colorbar(label='Amplitude')
plt.show()
```
