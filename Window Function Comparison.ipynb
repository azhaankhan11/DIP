```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a windowed signal using different window functions
t = np.linspace(0, 1, 1000)
signal = np.sin(2 * np.pi * 10 * t)
window_types = ['rectangular', 'hanning', 'hamming', 'blackman']
windows = [np.ones_like(signal), np.hanning(len(signal)), np.hamming(len(signal)), np.blackman(len(signal))]

# Compute the spectra of the windowed signals
freq = np.fft.fftfreq(len(signal), 1 / len(t))
spectra = [np.abs(np.fft.fft(signal * window)) for window in windows]

# Plot the spectra for different window functions
plt.figure(figsize=(10, 6))
for i, window_type in enumerate(window_types):
    plt.plot(freq, spectra[i], label=window_type)
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.title('Spectrum of Windowed Signals')
plt.legend()
plt.show()
```
