```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a time-domain signal
time = np.linspace(0, 1, 1000)
signal = np.sin(2 * np.pi * 10 * time) + 0.5 * np.sin(2 * np.pi * 25 * time)

# Compute the power spectrum using periodogram
freq, spectrum = plt.psd(signal, Fs=1000)

# Plot the power spectrum
plt.plot(freq, spectrum)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power Spectrum')
plt.title('Power Spectrum Estimation')
plt.show()
```
